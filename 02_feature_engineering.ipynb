```python
import re
from nltk.corpus import stopwords
```

```python
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv("D:\Restaurant Recommendation Project\data\clean_zomato.csv")
df.head()
```

```python
# Initialize a list of unique restaurant names from the 'name' column of the dataframe.
# Then, for each restaurant, calculate its mean rating and assign it to the 'Mean Rating' column
# for all rows corresponding to that restaurant.
restaurants = list(df['name'].unique())
df['Mean Rating'] = 0

for i in range(len(restaurants)):
    # Assign the mean rating to all rows corresponding to the current restaurant
    df['Mean Rating'][df['name'] == restaurants[i]] = df['rate'][df['name'] == restaurants[i]].mean()
```

```python
df.head()
```

```python
# Convert all text in the 'reviews_list' column to lowercase.
df["reviews_list"] = df["reviews_list"].str.lower()

# Display a random sample of 5 rows with 'reviews_list' and 'cuisines' columns.
df[['reviews_list', 'cuisines']].sample(5)
```

```python
# Access the 'reviews_list' column from the dataframe.
rest = df['reviews_list']

# Iterate through each review in the 'reviews_list' column.
for i in range(len(rest)):
    # Replace any non-alphanumeric characters (excluding spaces) in the review with a space.
    review = re.sub('[^a-z0-9]', ' ', rest[i])
```

```python
df['reviews_list']=review
df['reviews_list'].sample(5)
```

```python
# Import stopwords from the NLTK library 
STOPWORDS = set(stopwords.words('english'))

# The function splits the text into words and filters out those that are in the stopwords list.
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

# Apply the 'remove_stopwords' function to each entry in the 'reviews_list' column.
df["reviews_list"] = df["reviews_list"].apply(lambda text: remove_stopwords(text))
```

```python
# Define a function to remove URLs from a given text.
# The function uses a regular expression pattern to identify URLs starting with 'http' or 'https',
# or with 'www', and removes them from the text.
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Apply the 'remove_urls' function to each entry in the 'reviews_list' column.
df["reviews_list"] = df["reviews_list"].apply(lambda text: remove_urls(text))
```

```python
restaurant_names = list(df['name'].unique())
restaurant_names
```

```python
# Drop the columns 'address', 'rest_type', 'type', 'menu_item', and 'votes' from the DataFrame 'df'.
# The 'axis=1' argument specifies that columns (not rows) should be dropped.
df = df.drop(['address', 'rest_type', 'type', 'menu_item', 'votes'], axis=1)

# Display the first 5 rows of the updated DataFrame to verify the changes.
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize 'cost' and 'Mean Rating'
scaler = MinMaxScaler()
df['normalized_cost'] = scaler.fit_transform(df[['cost']])
df['normalized_rating'] = scaler.fit_transform(df[['Mean Rating']])
```

```python
# Create a new 'text' feature combining 'cuisines', 'location', 'reviews_list', 'cost', and 'rating'
df['text'] = df['cuisines'] + " " + df['location'] + " " + df['reviews_list'] + " " + \
             df['normalized_cost'].astype(str) + " " + df['normalized_rating'].astype(str)

# Check the first few rows to ensure the data looks good
df[['name', 'text']].head()
```

```python
# clean text data
df['text'] = df['text'].str.lower()  
df['text'] = df['text'].str.strip()  
df['text'] = df['text'].str.replace(',', '')  
df[['name', 'text']].head()
```

```python
# data save as clean_zomato.csv
df.to_csv("D:\Restaurant Recommendation Project\data\clean_data.csv", index=False)
```
